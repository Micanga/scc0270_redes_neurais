### Módulos necessários para realização do exercício

```python
import itertools

from matplotlib.colors import ListedColormap 
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions

import numpy as np
import pandas as pd
from re import match

from sklearn.metrics import precision_score, confusion_matrix
from sklearn.model_selection import learning_curve, train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```


## 1. Pré-Processamento<a id="sec:preproc"></a>

Nesta seção, iremos carregar os 2 conjuntos de dados e apresentá-los separadamente, visando facilitar seu entendimento.
Assim, a seção está dividida entre a [Seção 1.1](#sec:dataset1) e [Seção 1.2](#sec:dataset2), apresentando o pre-processamento sobre o dataset 1 e 2, respectivamente.

### 1.1. DATASET 1<a id="sec:dataset1"></a>

Primeiramente, iremos carregar o conjunto de dados "dados1.csv".

```python
# Carregando o conjunto de dados 1
X1, Y1 = [], []
with open('dados1.csv','r') as datafile:
    for line in datafile:
        if match('^\d+.\d+,\d+.\d+,\d+$',line) is not None:
            data = line.split(',')
            X1.append(np.array([float(d) for d in data[:-1]]))
            Y1.append(int(data[-1]))

# Definindo os vetores no formato de um vetor numpy
X1 , Y1 = np.array(X1), np.array(Y1)

# Construindo um dataframe do conjunto de dados para visualização
dataset_1 = np.column_stack((X1, Y1))
dataset_1 = pd.DataFrame(dataset_1)
dataset_1

# Construindo um gráfico para visualização do conjunto de dados
fig = plt.figure()
colors = ['red','blue']
plt.scatter(X1[:,0],X1[:,1], c=Y1, cmap= ListedColormap(colors))
plt.show()

# Dividindo o dataset 1 entre conjunto de treinamento e teste
X1_train, X1_test, Y1_train, Y1_test =\
    train_test_split(X1, Y1, test_size=0.33,stratify=Y1,random_state=1)
```

**OBS.:** Note que este conjunto de dados possui 2 parâmetros e a última coluna representa a classe da amostra. 
Ademais, existem duas classes possíveis para o conjunto de dados: Classe 1 e Classe 2; cuja amostra possui 511 elementos (sendo 342 para treinamento).

### 1.2. DATASET 2 <a id="sec:dataset2"></a>

Agora iremos carregar o conjunto de dados "dataset2.csv"

```python
# Carregando o conjunto de dados 2
X2, Y2 = [], []
with open('dados2.csv','r') as datafile:
    for line in datafile:
        if match('^\d+.\d+,\d+.\d+,\d+$',line) is not None:
            data = line.split(',')
            X2.append(np.array([float(d) for d in data[:-1]]))
            Y2.append(int(data[-1]))

# Definindo os vetores no formato de um vetor numpy
X2 , Y2 = np.array(X2), np.array(Y2)

# Construindo um dataframe do conjunto de dados para visualização
dataset_2 = np.column_stack((X2, Y2))
dataset_2 = pd.DataFrame(dataset_2)
dataset_2

# Construindo um gráfico para visualização do conjunto de dados
fig = plt.figure()
colors = ['red','blue']
plt.scatter(X2[:,0],X2[:,1], c=Y2, cmap= ListedColormap(colors))
plt.show()

# Dividindo o dataset 2 entre conjunto de treinamento e teste
X2_train, X2_test, Y2_train, Y2_test =\
    train_test_split(X2, Y2, test_size=0.33,stratify=Y2,random_state=1)
```

**OBS.:** Note que este conjunto de dados possui 2 parâmetros e a última coluna representa a classe da amostra. 
Ademais, existem duas classes possíveis para o conjunto de dados: Classe 1 e Classe 2; cuja amostra possui 196 elementos (sendo 131 para treinamento).

## 2. MLP e SVM - Arquitetura e Inicialização <a id="sec:mlpsvm"></a>

Nesta seção, serão apresentadas as arquiteturas e a inicialização dos métodos de classificação a serem utilizados que são uma <i>Multi Layer Perceptron</i> (MLP) e uma <i>Support Vector Machine</i> (SVM). 
Para esta apresentação, dividiu-se esta seção em duas subseções (Seções [2.1](#sec:mlp) e [2.2](#sec:svm)) para apresentação da MLP e da SVM, respectivamente. Ademais, esta seção realizará o treinamento de ambas técnicas, contudo a classificação em si será apresentada na [Seção 3](#sec:results).

### 2.1. Multi Layer Perceptron<a id="sec:mlp"></a>

Primeiramente, iremos inicializar uma MLP utilizando a <a href=https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html>classe implementada</a> no pacote Sklearn.

Como não iremos aplicar nenhuma técnica para se aproximar ou definir os hiper parâmetros da MLP, estes serão inicializados com valores que acreditamos ser suficientes para classificação.

Adicionalmente, serão construídas 2 redes iguais para que seja possível o treinamento de cada uma utilizando somente um dos dois conjunto de dados disponíveis (para futura avaliação).

```python
# Inicializando a MLP com uma arquitetura arbitrária
mlp = []
for i in range(2):
    mlp.append(MLPClassifier(hidden_layer_sizes=(4,2),\
                            activation='identity',\
                            solver='lbfgs',\
                            random_state=1))
```

Assim, construímos uma MLP cuja arquitetura segue:

1. **Arquitetura Geral:** (2:4:2:1)
    - **Número de entradas:** 2
    - **Número de camadas escondidas:** 2, com 4 e 2 neurônios, respectivamente.
    - **Número de saídas:** 1
    
2. **Parâmetros de Aprendizado:**
    - **Função de ativação:** identidade
    - **Otimizador:** optimizador da familia dos métodos quasi-Newton (considerando o tamanho dos datasets

Definida a arquitetura e inicializada a MLP, partiremos para o treinamento da rede.

```python
# Treinando as redes com os datasets 1 e 2, respectivamente
mlp_d1 = mlp[0].fit(X1_train,Y1_train)
mlp_d2 = mlp[1].fit(X2_train,Y2_train)
```

Agora, com as redes treinadas, habilitamos a realização de classificações com a MLP.

### 2.2. Support Vector Machine<a id="sec:svm"></a>

Primeiramente, iremos inicializar a SVM utilizando a <a href=https://scikit-learn.org/stable/modules/svm.html>classe implementada</a> no pacote Sklearn.

Para este exercício, utilizaremos a classe <i>Support Vector Classificator</i> (SVC) para classificação do dataset.
Descartou-se os classificadores Nu-SVC e o <i>Support Vector Regression</i> (SVR) devido as características dos datasets frente as formulações matemáticas de ambos tipos de SVM.

Adicionalmente, serão construídas: uma SVM Linear para o dataset 1 e uma SVM Não-Linear para o dataset 2.
Foi escolhida esta abordagem pois, pela visualização do dataset, sabe-se que ambas abordagens são capazes de contemplar a representação de classes e há interesse de se visualizar seus diferentes comportamentos.

```python
svm = []
# Definindo uma SVM Linear para o dataset 1
svm.append(SVC(kernel='linear'))

# Definindo uma SVM Nao-Linear para o dataset 2
svm.append(SVC(kernel='poly',gamma='auto'))
```

Definido os classificadores, partiremos para o treinamento de ambos classificadores.

```python
# Treinando os classificadores SVM
svm_d1 = svm[0].fit(X1_train,Y1_train)
svm_d2 = svm[1].fit(X2_train,Y2_train)
```

Agora, com os classificadores treinados, habilitamos a realização de classificações via SVM.

## 3. Resultados<a id="sec:results"></a>

Por meio de uma avaliação comparativa, iremos realizar a classificação dos conjuntos de dados utilizando a técnica MLP e SVM. Desta forma, esta seção está divida entre a coleta de resultados para o dataset 1 ([Seção 3.1](#sec:rd1)) e a coleta de resultados para o dataset 2 ([Seção 3.2](#sec:rd2)).

Para avaliação utilizaremos a análise de gráficos que apresentarão:
1. Região de decisão: a divisão feita pelo algoritmo do espaço de amostras
2. Curva de aprendizado: comportamento da curva de aprendizado do algoritmo (quanto a porcentagem de acerto a cada iteração de aprendizado).

### 3.1. Classficação para o dataset 1<a id='sec:red1'></a>

```python
# Classificando o conjunto de teste do dataset 1, calculando a precisão
# e imprimindo um gráfico com as regiões de decisões definidas
gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(10,8))

labels = ['MLP', 'SVM']
for clf, lab, grd in zip([mlp_d1, svm_d1],
                         labels,
                         itertools.product([0, 1], repeat=2)):
    
    # Imprimindo a precisão dos métodos
    print('Precisão da', lab,':',clf.score(X1_test,Y1_test))
    
    # Imprimindo o gráfico de regiões de decisão
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X1_test, y=Y1_test, clf=clf, legend=2)
    plt.title(lab)
    
    # Imprimindo o gráfico de curva de aprendizado
    train_sizes, train_scores, test_scores = \
        learning_curve(clf, X1_train, Y1_train,cv=3)

    test_scores_mean = 1- np.std(test_scores, axis=1)
    
    ax = plt.subplot(gs[grd[0]+1, grd[1]])
    ax.grid()
    ax.plot(train_sizes, test_scores_mean, 'o-', color="r",
                 label="Learning rate")
    ax.legend(loc="best")

plt.show()
```


### 3.1. Classficação para o dataset 2<a id='sec:red2'></a>

```python
# Classificando o conjunto de teste do dataset 2, calculando a precisão
# e imprimindo um gráfico com as regiões de decisões definidas
gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(10,8))

labels = ['MLP', 'SVM']
for clf, lab, grd in zip([mlp_d2, svm_d2],
                         labels,
                         itertools.product([0, 1], repeat=2)):
    
    # Imprimindo a precisão dos métodos
    print('Precisão da', lab,':',clf.score(X2_test,Y2_test))
    
    # Imprimindo o gráfico de regiões de decisão
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X2_test, y=Y2_test, clf=clf, legend=2)
    plt.title(lab)
    
    # Imprimindo o gráfico de curva de aprendizado
    train_sizes, train_scores, test_scores = \
        learning_curve(clf, X1_train, Y1_train,cv=3)

    test_scores_mean = 1- np.std(test_scores, axis=1)
    
    ax = plt.subplot(gs[grd[0]+1, grd[1]])
    ax.grid()
    ax.plot(train_sizes,test_scores_mean, 'o-', color="r",
                 label="Learning rate")
    ax.legend(loc="best")

plt.show()
```

## 4. Conclusão<a id="sec:conc"></a>

Realizando uma análise sobre os gráficos obtidos na seção de resultados, temos as seguintes conclusões:

1. Quanto a **precisão dos algoritmos**, nada se pode afirmar sobre vantagens e desvantagens entre os métodos avaliados, uma vez que, númericamente, a precisão de ambos é igual (i.e., ambos os métodos conseguem criar sub-espaços os quais representam bem os conjuntos de dados disponíveis).

2. Quando análisamos, precisamente, os **sub-espaços criados pelos algoritmos**, conseguimos apontar possíveis diferenças que podem trazer vantagens ou desvantagens para uma representação genérica do conjunto de dados.
    1. Quando comparamos a MLP e a SVM-Linear, temos um comportamento bem parecido para a criação de sub-espaços. Contudo, é possível perceber que a SVM tende a separar o espaço igualmente para ambas as classes, i.e., a reta que divide o sub-espaço encontra-se bem ao centro do meio espaço entre as amostras. Essa diferença, caso os dados sigam se dividindo uniformemente para cima e para baixo deste meio espaço, a representação será melhor generalizada pela SVM. Contudo, caso a distribuição de novas amostras não respeite essa divisão, nem a SVM, nem a MLP, possuem garantia de generalização. Note que, mesmo como classificadores similares, a SVM-Linear tende a apresentar uma melhor divisão do espaço para uma classificação genérica do que a MLP ao preço de um maior custo computacional (uma vez que ela funciona sobre a comparação one x one, apresentando uma pior escalabilidade com o aumento de amostras no conjunto de dados).
    2. Quando comparamos a MLP com a SVM-Polinomial, conseguimos visualizar que a SVM tende a representar a um sub-espaço que descreve o "viés de concentração" de cada amostra. Em outras palavras, podemos observar que, como a classe 2 parece seguir uma distribuição concentrada nas bordas do espaço descrito e a classe 1 uma distribuição concentrada na região sudoeste do espaço (distribuindo-se dentro de um circulo), a SVM apresenta esse viés quando traça a função separadora do espaço. Caso o conjunto de dados não represente bem o comportamento genérico de todas as amostras do conjunto, a SVM pode apresentar uma classificação pior que a MLP (que não é capaz de descrever sub-espaços polinomiais). Caso contrário, a SVM possuirá uma maior ou igual precisão da MLP ao custo de pior escalabilidade.
    
3. Quando avaliamos a **curva de aprendizado**, conseguimos concluir que a SVM, para esse conjunto de dados, consegue atingir uma melhor representação do espaço com uma menor quantidade de dados. Isso deve ocorrer pois a SVM já inicia seu aprendizado em alguma região do meio espaço entre as duas classes, que nesse conjunto de dados é bem definido e separa com uma distância significativa as duas classes. Já a MLP realiza um aprendizado iteração a iteração ajustando os pesos da rede.
